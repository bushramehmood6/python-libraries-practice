```python
planets["orbital_period"].fillna(planets["orbital_period"].mean(),inplace=True)
```

```python
planets["orbital_period"].isnull().sum()
```

```python
planets["mass"]=pd.to_numeric(planets["mass"],errors="coerce")
```

```python
planets["mass"]=planets["mass"].interpolate(method="linear",limit_direction="both")
```

```python
planets["mass"]=planets["mass"].fillna(planets["mass"].mean())
del planets["mass"]
```

```python
planets["distance"]=planets["distance"].interpolate(method="linear",limit_direction="both")
```
